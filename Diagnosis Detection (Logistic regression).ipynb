# Data cleaning

# Get dummy variables of label data

```python
diagnosis=pd.get_dummies(df["diagnosis"],drop_first=True)
```

```python
diagnosis.head()
```

# Delete unnecessary data

```python
df.drop(["id","diagnosis"],axis=1,inplace=True)
```

```python
df.head()
```

# Cancatenate Features and Label

```python
df=pd.concat([df,diagnosis],axis=1)
```

```python
df.head()
```

# Building Logistic Rrgression model

# Spliting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('M',axis=1),
                                               df["M"],test_size=0.22,
                                               random_state=101)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
predictions
```

# Confusion matrix

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
accuracy=accuracy_score(y_test,predictions)
```
